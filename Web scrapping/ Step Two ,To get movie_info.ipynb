```python
Movie_info =[]
for ids in movie_ID_array:
    # Prepare the URL by replacing spaces with '+'
    search_url =f"https://www.imdb.com/title/{ids}/"
    print(search_url,"ii")
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.9999.99 Safari/537.36'
    }
    response = requests.get(search_url, headers=headers)
    if response.status_code == 200:
       soup = bs(response.text, 'html.parser')

       Movie_element=soup.find('span', class_='sc-afe43def-1 fDTGTb')
       Movie_name = Movie_element.text  if Movie_element else 'N/A'
       print("Movie Name:",Movie_name)
        
       date_element=soup.find_all(class_="ipc-link ipc-link--baseAlt ipc-link--inherit-color")
       if len(date_element) >= 2:
          date_element = date_element[-2]
          date = date_element.text
       else:
          date = 'N/A'
       print("Date:",date) 
        
       rating_element =soup.find('span', class_='sc-bde20123-1 iZlgcd')
       rating = rating_element.text if rating_element else 'N/A'
       print("Rating:",rating)
        
       genre_elements = soup.find_all(class_="ipc-chip__text")[0:-1]
       genres = [genre.text for genre in genre_elements] if genre_elements else []
       genre_list = ",".join(genres)
       print("Genres:", genre_list)
        
       cast_element = soup.find_all(class_="sc-bfec09a1-1 fUguci")[0:-1]
       cast = [genre.text for genre in cast_element] if cast_element else []
       cast_list = ",".join(cast)
       print("cast:",cast_list)

       movie_info = {
            "Movie Name":Movie_name,
            "Rating":rating,
            "Date":date,
            "Genres":genre_list,
            "cast":cast_list,
            "Imdb _id":ids
        }

       print("-------")

       Movie_info.append(movie_info)
    else:
        print("status:",response.status_code)
        Movie_info.append(movie_info)
        
        



       
       
       
       
    
```

```python
csv_file_path = "movie_info2.csv"

# Define the header for the CSV file (column names)
header = ["Movie Name","Imdb_id", "Rating", "Release Date", "Genres", "Cast"]

# Write the data to the CSV file
with open(csv_file_path, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    
    # Write the header row
    writer.writerow(header)
    
    # Write the data for each movie
    for movie in Movie_info:
        row = [movie["Movie Name"],movie["Imdb _id"], movie["Rating"], movie["Date"], movie["Genres"], movie["cast"]]
        writer.writerow(row)

print("Data has been written to", csv_file_path)





```
