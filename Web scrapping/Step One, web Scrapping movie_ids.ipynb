```python
## Here is the logic Code to retrive Whole Ids
```

```python
movie_titles_array
# Create an empty dictionary to store the IMDb search results
imdb_ids ={}

# Loop through each movie title
for movie_name in movie_titles_array:
    # Prepare the URL by replacing spaces with '+'
    search_url = f"https://www.imdb.com/find?q={movie_name.replace(' ', '+')}"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.9999.99 Safari/537.36'
    }
    response = requests.get(search_url, headers=headers)
    if response.status_code == 200:
       soup = bs(response.text, 'html.parser')
       link = soup.find('a', class_='ipc-metadata-list-summary-item__t')
       # imdb_id = link['href'].split('/')[2] 
       # print(imdb_id,"imdb_id")
       if link:
            imdb_id = link['href'].split('/')[2]
            print("id-->",imdb_id,movie_name)
       else:
           imdb_id = "null"
       imdb_ids[movie_name] = imdb_id  
print(imdb_ids)
          
        
    
       
                
           

      


```

```python
## Here is the to download a whole ids as csv file for fuether process
```

```python
csv_file_path = "imdb_ids.csv"

# Open the CSV file for writing
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # Write the data from the 'imdb_ids' dictionary to the CSV file
    writer.writerow(['Movie Name', 'IMDb ID'])  # Write header row
    for movie_name, imdb_id in imdb_ids.items():
        writer.writerow([movie_name, imdb_id])

print(f'Data has been saved to {csv_file_path}')
```
